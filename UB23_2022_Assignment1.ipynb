Q7: Obtain Name, Language and percentage in Africa that have 82% or more speakers of a single non-English language. Do **NOT** use subquery/CTE.


````
%%read_sql df_q7
SELECT country.Name,countrylanguage.Language, SUM(Percentage) AS percentage FROM country 
JOIN countrylanguage ON country.code = countrylanguage.CountryCode
WHERE countrylanguage.Language != 'English'
AND continent = 'Africa'
GROUP BY country.Name,countrylanguage.Language 
HAVING SUM(Percentage)>=82
ORDER BY SUM(Percentage) DESC
;
````
